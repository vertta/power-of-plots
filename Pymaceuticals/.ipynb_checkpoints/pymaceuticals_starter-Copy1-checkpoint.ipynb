```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head()
```

```python
# Checking the number of mice.
len(study_data_complete["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = study_data_complete[study_data_complete.duplicated(subset=('Mouse ID','Timepoint'),keep=False)]['Mouse ID'].unique()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
study_data_complete[study_data_complete["Mouse ID"].isin(duplicates)]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleanDataFrame = study_data_complete
cleanDataFrame.drop(study_data_complete[study_data_complete["Mouse ID"].isin(duplicates)].index, inplace=True)

cleanDataFrame.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleanDataFrame["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mumor Volume (mm3)ean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drugGroup = cleanDataFrame.groupby("Drug Regimen")
dgMean  = drugGroup["Tumor Volume (mm3)"].mean()
dgMedian = drugGroup["Tumor Volume (mm3)"].median()
dgVariance = drugGroup["Tumor Volume (mm3)"].var()
dgStd = drugGroup["Tumor Volume (mm3)"].std()
dgSem = drugGroup["Tumor Volume (mm3)"].sem()
summary_table = pd.DataFrame(
    {"Mean Tumor Volume": dgMean,
     "Median Tumor Volume":dgMedian,
     "Tumor Volume Variance": dgVariance,
     "Tumor Volume Std. Dev.":dgStd,
      "Tumor Volume Std. Err.": dgSem
            
})
summary_table.name='Tumor Volume (mm3)'

#summarytest = drugGroup[[summaryDF]]
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_table=cleanDataFrame.groupby("Drug Regimen").agg({
    "Tumor Volume (mm3)": ["mean","median","var","std","sem"]
})
# Using the aggregation method, produce the same summary statistics in a single line

summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

greatestTimePoint=cleanDataFrame.groupby("Mouse ID")["Timepoint"].max()
greatestTimePoint = greatestTimePoint.reset_index()
greatestTimePoint.head()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_data = greatestTimePoint.merge(cleanDataFrame,on=["Mouse ID","Timepoint"], how="left")
merged_data.head()
```

```python
merged_data["Drug Regimen"].unique()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatment_list:
      # Locate the rows which contain mice on each drug and get the tumor volumes
    treatmentfinal = merged_data.loc[merged_data["Drug Regimen"]== drug,"Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol_list.append(treatmentfinal)
    
    # Determine outliers using upper and lower bounds
    quartiles = treatmentfinal.quantile([0.25,0.50,0.75])
    lowervalue = quartiles[0.25]
    uppervalue = quartiles[0.75]
    midvalue = uppervalue  - lowervalue 
    lowerbound = lowervalue -( 1.5 * midvalue)
    upperbound = uppervalue + (1.5 * midvalue)
    outliers  = treatmentfinal.loc[(treatmentfinal < lowerbound) | (treatmentfinal < lowerbound)] 
    
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
finalplot = dict(markerfacecolor="red", markersize=12)
plt.boxplot(tumor_vol_list, labels=treatment_list, flierprops=finalplot)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```
